```python
sent1 = "Seorang putri rupawan. Menatap kerlip bintang tak berjumlah. Lalu berkata perlahan pada ayahandanya"
sent2 = "*Pemenang I Sayembara Novel DKJ 2008*""Di ujung sabar ada perlawanan."""
sent3 = "-- Pemenang Sayembara Menulis Novel Dewan Kesenian Jakarta 2012 --“...ditulis dengan teknik penceritaan yang intens,"
sent4 = """Selamat, Maria!"" seru Guntur, susah payah mendekati gadis itu. ""Mainmu bagus sekali!"""
sent5 = "“Jam tiga dini hari, sweter, dan jalanan yang gelap dan sepi .... Ada peta, petunjuk; dan Jakarta"

sentences = {
    'raw': [sent1, sent2, sent3, sent4, sent5]
}

df = pd.DataFrame(sentences)
df
```

```python
df["lower"] = [desc.lower() for desc in df["raw"]]
df
```

```python
df["cleaned"] = [cleaning.clean(desc) for desc in df["lower"]]
df
```

```python
df["stopword_removed"] = [cleaning.remove_stopword(desc) for desc in df["cleaned"]]
df
```

```python
df["stemmed"] = [cleaning.stem(desc) for desc in df["stopword_removed"]]
df
```

```python
df["tokenized"] = df["stemmed"].str.split()
df.to_csv('sample_preprocessed.csv', index=False)
df
```
